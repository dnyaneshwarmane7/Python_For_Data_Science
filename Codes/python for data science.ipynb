```python
for i in range(0, len(cars_data['Price']), 1):
    if (cars_data['Price'][i] <= 8450):
        cars_data['Price_Class'][i] = 'low'
    elif (cars_data['Price'][i] > 11950):
        cars_data['Price_Class'][i] = 'High'
    else:
        cars_data['Price_Class'][i] = 'Medium'
```

```python
cars_data['Price_Class'].value_counts()
```

```python
cars_data.insert(11, 'Age_Converted',0)
cars_data.insert(12, 'Km_per_month', 0)
```

```python
def c_convert(val1, val2):
    val_converted = val1/ 12
    ratio = val2/val1
    return [val_converted, ratio]
```

```python
cars_data['Age_Converted'], cars_data['Km_per_month'] = c_convert(cars_data['Age'], cars_data['KM'])
```

## Dealing with missing values

```python
cars_data = pd.read_csv('Toyota.csv', index_col=0, na_values=['??', '###','????'])
```

```python
## Creating copies
cars_data2 = cars_data.copy()
cars_data3 = cars_data.copy()
```

```python
## Identifying missing values
```

```python
cars_data2.isna().sum()
#cars_data2.isnull().sum()
```

```python
missing = cars_data2[cars_data2.isnull().any(axis=1)]
missing
```

## Approaches to fill the missing values
1. Fill the missing value by mean/ median, in case of numerical value
2. Fill the missing values with the class which has maximum count, in case of categorical variable

### Imputing missing values

```python
cars_data2.describe()
```

```python
## Imputing missing values of 'Age'

# Calculating mean value of Age variable

cars_data2['Age'].mean()
```

```python
cars_data2['Age'].fillna(cars_data2['Age'].mean(), inplace = True)
```

```python
## Imputing missing values of 'KM'

# Calculating median value of KM variable
cars_data2['KM'].median()
```

```python
cars_data2['KM'].fillna(cars_data2['KM'].median(), inplace = True)
```

```python
## Imputing missing values of 'HP'

# Calculating mean value of HP variable
cars_data2['HP'].mean()
```

```python
cars_data2['HP'].fillna(cars_data2['HP'].mean(), inplace = True)
```

```python
## Check for missing values from dataframe after filling values
cars_data2.isnull().sum()
```

```python
## Imputing missing values of 'FuelType'
# Get frequency

cars_data2['FuelType'].value_counts().index[0]
```

```python
cars_data2['FuelType'].fillna(cars_data2['FuelType'].value_counts().index[0],
                             inplace=True)
```

```python
## Imputing missing values of 'MetColor'

cars_data2['MetColor'].mode()
```

```python
cars_data2['MetColor'].fillna(cars_data2['MetColor'].mode()[0],
                             inplace=True)
```

```python
## Check for missing values from dataframe after filling values
cars_data2.isnull().sum()
```

## Imputing missing values using lambda functions

```python
cars_data3 = cars_data3.apply(lambda x:x.fillna(x.mean()) \
                              if x.dtype == 'float' else \
                              x.fillna(x.value_counts().index[0]))
```

```python
## Check for missing values from dataframe after filling values
cars_data3.isnull().sum()
```
