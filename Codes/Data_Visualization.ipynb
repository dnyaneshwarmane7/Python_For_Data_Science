## Scatter Plot

What is a scatter plot?
1. A scatter plot is a set of points that represents the values obtained for two different variables plotted on a horizontal and vertical axes

When to use scatter plots?
1. Scatter plots are used to convey the relationship between two numerical variables
2. Scatter plots are sometimes called correlationplots because they show how two variables arecorrelated

```python
plt.scatter(cars_data['Age'], cars_data['Price'], c='red')
plt.title('Scatter plot of Price vs Age of Cars')
plt.xlabel('Age(Months)')
plt.ylabel('Price(Euros)')
plt.show()
```

#### Conclusion - The price of the car decreases as age of the car increases

### Scatter plot with Seaborn Library

Scatter plot of Price vs Age by customizing the appearance of markers

```python
sns.set(style = 'darkgrid')
sns.regplot(x = cars_data['Age'], y=cars_data['Price'], marker = '*')
plt.title('Scatter plot of Price vs Age of Cars')
plt.xlabel('Age(Months)')
plt.ylabel('Price(Euros)')
plt.show()
```

### Scatter plot of Price vs Age by FuelType

Using hue parameter, including another variable to show the fuel 
types categories with different colors

```python
sns.lmplot(x='Age', y='Price', data=cars_data,
          fit_reg=False, hue='FuelType',
          legend=True, palette='Set1')
plt.title('Scatter plot of Price vs Age by FuelType')
```

## Histogram

What is a histogram?
1. It is a graphical representation of data usingbars of different heights
2. Histogram groups numbers into ranges andthe height of each bar depicts the frequency of each range or bin

When to use histograms?
1. To represent the frequency distribution of mnumerical variables


```python
plt.hist(cars_data['KM'], color='green', edgecolor = 'white' , bins= 5)
plt.title("Histogram of Kilometer")
plt.xlabel("Kilometer")
plt.ylabel('Frequency')
plt.show()
```

#### Conclusion : Frequency distribution of kilometre of the cars shows that most of the cars have travelled between 50000 â€“ 100000 km and there are only few cars with more distance travelled


```python
sns.distplot(cars_data['Age'],  bins= 5)
plt.title('Histogram of Age')
plt.show()
```

## Bar plot

What is a bar plot?
1. A bar plot is a plot that presents categorical data with rectangular bars with lengths proportional to the counts that they represent

When to use bar plot?
1. To represent the frequency distribution of categorical variables
2. A bar diagram makes it easy to compare sets of data between different groups

```python
cars_data['FuelType'].value_counts()
```

```python
counts = [1277, 144, 15]
fuelType = ('Petrol','Diesel', 'CNG')
index = np.arange(len(fuelType))
```

```python
plt.bar(index, counts, color = ['red','blue','cyan'])
plt.title('Bar plot of fuel types')
plt.xlabel('Fuel Types')
plt.ylabel('Frequency')
plt.xticks(index, fuelType, rotation = 90)
plt.show()
```

#### Conclusion - Bar plot of fuel type shows that most of the cars have petrol as fuel type


### With Seaborn library

## Grouped bar plot

Grouped bar plot of FuelType and Automatic

## Box-whiskers plot and Histogram

1. Letâ€™s plot box-whiskers plot and histogram on the same window
2. Split the plotting window into 2 parts then add create two plots

```python
f, (ax_box, ax_hist) = plt.subplots(2, gridspec_kw={'height_ratios':(0.30, 0.70)})
sns.boxplot(cars_data['Price'], ax= ax_box)
sns.distplot(cars_data['Price'], ax= ax_hist, kde=False)
```
